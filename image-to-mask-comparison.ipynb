```python
def show_img(im, figsize=None, ax=None, alpha=None):
    if not ax: fig,ax = plt.subplots(figsize=figsize)
    ax.imshow(im, alpha=alpha)
    ax.set_axis_off()
    return ax
```

```python
TRAIN_DN = 'train/images'
MASKS_DN = 'train/masks'
sz = 128
bs = 64
```

```python
ims = [open_image(Path(TRAIN_DN)/Path(image)) for image in os.listdir('train/images')[:16]]
im_masks = [open_image(Path(MASKS_DN)/Path(mask)) for mask in os.listdir('train/masks')[:16]]
```

```python
fig, axes = plt.subplots(4, 4, figsize=(9, 6))
for i,ax in enumerate(axes.flat):
    ax = show_img(ims[i], ax=ax)
    show_img(im_masks[i][...,0], ax=ax, alpha=0.5)
plt.tight_layout(pad=0.1)
```
